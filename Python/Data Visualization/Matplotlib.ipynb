# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','Mar','Apr','May','June']
traffic =[150, 2, 200, 220, 250, 310]
plt.plot(months, traffic)
plt.show()
```

```python
#sample
year=[2004,2005,2006,2007]
birth_count=[23456,456789,456789,345678]
plt.plot(year,birth_count)
plt.xlabel('year')
plt.ylabel('birthcount')
plt.title("Birth_Count_by_Year")
plt.show()
```

```python
#adding styles
year=[2004,2005,2006,2007]
birth_count=[23456,456789,456789,345678]
plt.plot(year,birth_count, marker='^', markersize=30, color ="cyan", linestyle='--',linewidth=3)
plt.xlabel('year')
plt.ylabel('birthcount')
plt.title("Birth_Count_by_Year")
plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun']
product_a_revenue=[35,45,45,67,24,87]
product_b_revenue=[35,97,46,97,97,23]
plt.plot(months, product_a_revenue, marker='o', linestyle='-', color='blue', label='Product A')
plt.plot(months, product_b_revenue, marker='s', linestyle='--', color='red', label='Product B')
plt.xlabel('Month')
plt.ylabel('Monthly Revenue (in $1000)')
plt.title('Monthly Revenue Comparison')
plt.legend()
plt.show()
```

```python
#cricket
import random
over=[i for i in range(1,21)]
runrate_a=[21,5,23,33,28,17,7,16,13,11,33,28,17,7,16,9,18,14,19,12]
runrate_b=runrate_a[::-1]
plt.plot(over, runrate_a, marker='o', linestyle='-', color='blue', label='Team A')
plt.plot(over, runrate_b, marker='s', linestyle='--', color='red', label='Team B')
plt.xlabel('Over')
plt.ylabel('Run_Rate')
plt.title('Team A and B Comparison')
plt.legend()
plt.show()
```

## Saving plot as an image

```python
plt.plot(months, traffic, marker='o', linestyle='--', color='g') #custom appearance
plt.xlabel('Month')
plt.ylabel('Monthly Traffic (in Thousands)')
plt.title('Monthly Website Traffic')
plt.savefig('pythonlinechart') #to save the file as image
plt.grid()
plt.show()
```

## bar chart

```python
categories=['Category A','Category B','Category C','Category D']
values=[20,35,30,25]
plt.bar(categories, values, color=['red','g','black'], width=0.8)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

```python
plt.barh(categories, values, color=['red','g','black'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```
