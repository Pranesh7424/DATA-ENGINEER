```python
#sample
year=[2004,2005,2006,2007]
birth_count=[23456,456789,456789,345678]
plt.plot(year,birth_count)
plt.xlabel('year')
plt.ylabel('birthcount')
plt.title("Birth_Count_by_Year")
plt.show()
```

```python
#adding styles
year=[2004,2005,2006,2007]
birth_count=[23456,456789,456789,345678]
plt.plot(year,birth_count, marker='^', markersize=30, color ="cyan", linestyle='--',linewidth=3)
plt.xlabel('year')
plt.ylabel('birthcount')
plt.title("Birth_Count_by_Year")
plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun']
product_a_revenue=[35,45,45,67,24,87]
product_b_revenue=[35,97,46,97,97,23]
plt.plot(months, product_a_revenue, marker='o', linestyle='-', color='blue', label='Product A')
plt.plot(months, product_b_revenue, marker='s', linestyle='--', color='red', label='Product B')
plt.xlabel('Month')
plt.ylabel('Monthly Revenue (in $1000)')
plt.title('Monthly Revenue Comparison')
plt.legend()
plt.show()
```

```python
#cricket
import random
over=[i for i in range(1,21)]
runrate_a=[21,5,23,33,28,17,7,16,13,11,33,28,17,7,16,9,18,14,19,12]
runrate_b=runrate_a[::-1]
plt.plot(over, runrate_a, marker='o', linestyle='-', color='blue', label='Team A')
plt.plot(over, runrate_b, marker='s', linestyle='--', color='red', label='Team B')
plt.xlabel('Over')
plt.ylabel('Run_Rate')
plt.title('Team A and B Comparison')
plt.legend()
plt.show()
```

## Stack plot

```python
import matplotlib.pyplot as plt
quarters=['Q1','Q2','Q3','Q4',]
electronics=[10000,12000,11000,10500]
Clothing=[5000,6000,7500,8000]
home_app=[3000,7500,8200,9000]
plt.figure(figsize=(10,6))
plt.stackplot(quarters,electronics,Clothing,home_app,labels=['Electronics','Clothing','Home Appliances'])
plt.xlabel('quarters')
plt.ylabel('Sales($)')
plt.title('Product Category')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y = np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2+y**2))
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)
fig,axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].plot(x,y1,'r')
axs[0,0].set_title('Sine Wave')
axs[0,1].plot(x,y2,'b')
axs[0,1].set_title('Cosine Wave')
axs[1,0].plot(x,y1,'r')
axs[1,0].set_title('Tangent Wave')
axs[1,1].plot(x,y2,'b')
axs[1,1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```
