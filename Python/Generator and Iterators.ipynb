## Generator and Iterators 

### Iterator
#### An iterator is an object that allows you to traverse through all the elements of a collection (like lists or tuple) one at a time

### key Points
#### __iter()__ and __next__() methods
#### raises stopIteration when no more data is available
#### used in loops like for loop

```python
class MyNumbers:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num <=5:
            x=self.num
            self.num +=1
            return x
        else:
            raise StopIteration
myclass=MyNumbers()
myiter=iter(myclass)
for x in myiter:
    print(x)
```

```python
class sq_numbers:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num <=10:
            x=self.num **2
            self.num +=1
            return x
        else:
            raise StopIteration
sq=sq_numbers()
iiter=iter(sq)
for x in iiter:
    print(x)
```

## Generator
#### A generator is  a special type of iterator that is defined using a function and the 'yield' keyword

### keypoints
#### uses yield instead of returns
#### Generates values lazily one at a time
#### More memory efficient than normal functions

```python
def my_generator():
    for i in range(1,6):
        yield i
gen=my_generator()
for value in gen:
    print(value)
```

### iterator vs Generator
#### iterator: requires class with __iter__() and __next__()
#### generator: uses function with yield keyword
#### iterator: more code, manual implementation
#### genrator: simple and concise implementation
#### iterator: slightly more control
#### genrator: more memory efficient

```python
def my_gen():
    for i in range(3,50,3):
        yield i
gen=my_gen()
for i in gen:
    print(i)
```
