## if else with for loop

```python
x=[1,4,5,6]
if x:
    for data in x:
        print(data)
    print("Data over...")
else:
    print("No data")
```

## nested if else

```python
num = 15
if (num>10 or 10>num):
    if (num>20):
        print("number is positive")
    else:
        print("2nd division")
else:
    print("Fail")
```

## For loop with if else
