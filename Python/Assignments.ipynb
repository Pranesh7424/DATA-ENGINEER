### 1.Matrix transpose using with and without built in function

```python
#without built-in
matrix=[[1,2,3],[4,5,6],[7,8,9]]
transpose=[]
for i in range(len(matrix[0])):
    row=[]
    for j in range(len(matrix)):
        row.append(matrix[j][i])
    transpose.append(row)
print(transpose)
```

```python
#with built-in
matrix=[[1,2,3],[4,5,6],[7,8,9]]
transpose=list(zip(*matrix))
print(transpose)
```

### 1.Why the local variable not callable in outside function

```python
#because it exists only inside function.
def test():
    x=10
    print(x)
test()
print(x)
```

```python
#want outside access,use return:
def test():
    x=10
    return x
value=test()
print(value)
```

### 2.where do you use map,filter and reduce

```python
#map-apply same operation to all elements
nums=[1,2,3,4]
result=list(map(lambda x:x*2,nums))
print(result)
```

```python
#filter-select elements based on condition
nums=[1,2,3,4,5]
result=list(filter(lambda x:x%2==0,nums))
print(result)
```

```python
#reduce-single output from list
from functools import reduce
nums=[1,2,3,4]
result=reduce(lambda x,y:x+y,nums)
print(result)
```

### 1.Calc with class and using error handling

```python
class Calc:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def add(self):
        return self.num1 + self.num2
    def sub(self):
        return self.num1 - self.num2
    def mul(self):
        return self.num1 * self.num2
    def div(self):
        return self.num1 / self.num2
while True:
    try:
        print("\n========== Simple Calculator ==========")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")
        print("5. Exit")
        print("=======================================")
        choice = int(input("Select an option (1-5): "))
        if choice == 5:
            print("Thank you for using the Calculator ðŸ‘‹")
            break
        if choice not in [1, 2, 3, 4]:
            raise ValueError("Invalid option selected!")
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        calc = Calc(num1, num2)
        if choice == 1:
            result = calc.add()
        elif choice == 2:
            result = calc.sub()
        elif choice == 3:
            result = calc.mul()
        elif choice == 4:
            result = calc.div()
        print(f"Result: {result}")
        print("Operation completed successfully!")
    except Exception as e:
        print(f"[Error] {e}")
```

### 1.Unit testing

```python
class ShoppingCart:
    def __init__(self):
        self.items = {}
    def add_item(self, name, price, quantity=1):
        if price <= 0 or quantity <= 0:
            raise ValueError("Invalid price or quantity")
        self.items[name] = (price, quantity)
    def remove_item(self, name):
        if name not in self.items:
            raise KeyError("Item not found")
        del self.items[name]
    def total_price(self):
        total = 0
        for price, quantity in self.items.values():
            total += price * quantity
        return total
    def apply_discount(self, percent):
        if percent < 0 or percent > 100:
            raise ValueError("Invalid discount")
        return self.total_price() * (1 - percent / 100)
```

```python
import unittest
class TestShoppingCart(unittest.TestCase):
    def setUp(self):
        self.cart = ShoppingCart()
        self.cart.add_item("Laptop", 50000, 1)
        self.cart.add_item("Mouse", 500, 2)
    def test_total_price(self):
        self.assertEqual(self.cart.total_price(), 51000)
    def test_apply_discount(self):
        self.assertEqual(self.cart.apply_discount(10), 45900)
    def test_invalid_discount(self):
        with self.assertRaises(ValueError):
            self.cart.apply_discount(150)
    def test_remove_item(self):
        self.cart.remove_item("Mouse")
        self.assertEqual(self.cart.total_price(), 50000)
    def test_remove_non_existing_item(self):
        with self.assertRaises(KeyError):
            self.cart.remove_item("Keyboard")
unittest.main(argv=[''],verbosity=3,exit=False)
```
