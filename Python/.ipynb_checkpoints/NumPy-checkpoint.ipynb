## Basic Data Types

```python
a=np.arange(10,dtype=float)
print(a.dtype)
print(type(a))
print(a)
```

## Indexing & Slicing

```python
a=np.arange(10)
a[2]=5 #assign #if matrix [2,1] rows and column based
print(a)
print(a[1:9:2])
a[5:] =10
a
```

```python
b=np.arange(5)
a[5:] = b[::-1] #takes value from reverse and assign into a starts from index 5
b
a
```

## Copies and Views

```python
#copies the data and mem not address
# only using slice operator
#if we use assign means same address
a=np.arange(10)
a
b=a[::2]
b
c=a[::3]
c
print(id(a))
print(id(b))
print(id(c))
```

```python
np.may_share_memory(a,b) #module name
```

```python
np.may_share_memory(b,c)
```

```python
b[0]=10
b
```

```python
a #changes happen from b to a
```

```python
#copies the data, not address
#if we want data slicing and different address
a=np.arange(10)
print(a)
c=a[::2].copy()
c
```

```python
print(np.may_share_memory(a,c))
print(id(a))
print(id(c))
c[0]=5 #changes not happen because it copy
print(c) 
a
```
