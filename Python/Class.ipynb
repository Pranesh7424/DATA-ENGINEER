## init

```python
class member:
    def __init__(self, name, age, position=None):
        self.name=name
        self.age=age
        self.position=position
    def hi(self, name, age):
        self.name = name
        self.age=age
test = member("Joe",45,"officer")
test1=member("Jan",49)
print(test.name)
print(test.age)
print(test.position)
print(test1.name)
print(test1.age)
```

## python object

```python
class member:
    def __init__(self, name, age, position):
        self.name=name
        self.age=age
        self.position=position
    def testfun(self):
        print("Hello my name is"+ self.name)
n1=member("Roger",34,"Manger")
print(n1.name)
print(n1.age)
print(n1.position)
n1.testfun()
```

## self parameter

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def greet(xyz):
        print("Hello my Name is "+ xyz.name)
p=Person("John",33)
p.greet()
```

## inheritance

```python
class a:
    print("This is parent a")
class b(a):
    print('This is b inherit from a' )
```

```python
class Sports:
    def __init__(self, color, brand, cost):
        self.__color = color
        self.__brand = brand
        self.__cost = cost
    def setcolor(self, color):
        self.__color = color
    def getcolor(self):
        return self.__color
    def setcost(self, cost):
        self.__cost = cost
    def getcost(self):
        return self.__cost
class Shoes(Sports):
    def __init__(self, model, color, brand, cost):
        super().__init__(color, brand, cost)
        self.model = model
sh1 = Shoes("Running", "Black", "Nike", 45000)
print("Cost:", sh1.getcost())
```
