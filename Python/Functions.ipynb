## hcf or gcd

```python
import math
def hcf(n1,n2):
    return math.gcd(n1,n2)
result=hcf(5,6)
print(result)
```

## filter()

```python
def check_positive(num):
    if num>0:
        return num
lst=[10,-20,3]
lstpos=list(filter(check_positive,lst))
lstneg=list(filter(lambda x:x<0,lst))
lstodd=list(filter(lambda n:n%2!=0,lst))
print(lstpos)
print(lstneg)
print(lstodd)
```

## map

```python
num=[1,2,3,4]
def hallsquare(num):
    return (num**2)
squared=list(map(hallsquare,[1,2,3,4]))
squared1=list(map(lambda x:x**2, [1,2,3,4]))
print(squared)
print(squared1)
```

## keyword argument

```python
def lst(**kwargs):
    if kwargs:
        print("Contact {0},{1},{2}".format(kwargs['name'],kwargs['mob'],kwargs['city']))
lst(name="Gopal",mob=9786543245,city="kerala")
```

## Arbitary argument

```python
def lst(*names):
    print(names)
    for lst2 in names:
        print("Name {0}".format(lst2))
lst([5,6,7],'sam','ram','jam',3.5)
```
